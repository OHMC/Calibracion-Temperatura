```python
import metricas as mt
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

### 60 cuadras

```python
wrf_60_A = pd.read_pickle('../pickles/60cuadras_wrf_A')
wrf_60_B = pd.read_pickle('../pickles/60cuadras_wrf_B')
wrf_60_C = pd.read_pickle('../pickles/60cuadras_wrf_C')
wrf_60_D = pd.read_pickle('../pickles/60cuadras_wrf_D')
```

```python
base = '/home/sagus/Development/Calibracion-Temperatura/datasets/'
filename = 'omixom_60Cuadras.csv'
path = f"{base}{filename}"
```

```python
aws_60 = mt.fromCSVtoDF(path)
```

```python
dfTmp_60_A = mt.getAWSvsWRF(wrf_60_A, aws_60, '60', 'A')
dfTmp_60_B = mt.getAWSvsWRF(wrf_60_B, aws_60, '60', 'B')
dfTmp_60_C = mt.getAWSvsWRF(wrf_60_C, aws_60, '60', 'C')
dfTmp_60_D = mt.getAWSvsWRF(wrf_60_D, aws_60, '60', 'D')
```

```python
data_60_A, rmse_60_A, mae_60_A = mt.getMetricas(dfTmp_60_A, '60', 'A')
data_60_B, rmse_60_B, mae_60_B = mt.getMetricas(dfTmp_60_B, '60', 'B')
data_60_C, rmse_60_C, mae_60_C = mt.getMetricas(dfTmp_60_C, '60', 'C')
data_60_D, rmse_60_D, mae_60_D = mt.getMetricas(dfTmp_60_D, '60', 'D')
```

```python
data_60_B
```

```python
ax = plt.gca()

data_60_A.plot(kind='line', ax=ax)
data_60_B.plot(kind='line', ax=ax)
data_60_C.plot(kind='line', ax=ax)
data_60_D.plot(kind='line', ax=ax)

plt.xlabel('Hora (UTC)')
plt.ylabel('Error (WRF - AWS) K')
plt.title('Grafico de errores de temperatura en Camino a 60 cuadras')
```
